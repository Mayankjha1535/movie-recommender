```python
def PCA(i1,i2):
    i1_a=i1-i1.mean()
    i2_a=i2-i2.mean()
    return np.sum(i1_a*i2_a)/np.sqrt(np.sum(i1_a**2)*np.sum(i2_a**2))
 
```
